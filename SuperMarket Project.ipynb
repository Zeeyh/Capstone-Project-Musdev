### Calculate and report the average total spending between females and males in superstores, and the average total spending between Londoners and non-Londoners in express stores. Discuss your findings

```python
# Average total spend between male and females

df[["superstore.total.spend", "gender", "income"]].groupby("gender").mean().astype(int)
```

- Females have an average total spending of £39,640
- Males have an average of £37,937

The total spending of females on average at superstores is slightly above that of the males, which makes sense since the females that spent at the superstores had slightly more income on average, than their male counterparts

```python
# Average total spend for Londoners

df[["Region","express.total.spend","income"]].groupby("Region").mean().astype(int)
```

- Londoners have an average total spending of £1,244. This comes third in line to Central England and South West who have an average total spend of £1,376 and £1,228 respectively.

Not much can be inferred from this except that Londoners were the 3rd highest spenders on average at the express superstores. Other regions save for Central England and South West, seemed to have low total spending in express stores on average.
